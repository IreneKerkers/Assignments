## Graphs

To make the graphs we need to make selections of the data. Please use "dplyr" (commands "select" and "filter") and the "pipe". The first assignment is to make a scatterplot using "ggplot". The graph contains data for all countries for the year 2014. Therefore make a dataframe with dplyr first, called "graph1" with the required data.
Then, make a scatterplot with "Cost" on the x-axis and "LE" on the y-axis. You have to add a regression line with a shaded area for the confidence interval.

```r
graph1 <- data.frame(oecd) %>% 
        select(Country, Year, Cost, LE) %>% 
        filter(Year == 2014)
graph1
```

```r
oecd.scatter <- ggplot(graph1, aes(x= Cost, y= LE)) + 
                geom_point(shape=16) + 
                stat_smooth(method = "lm", col = "purple") +
                theme(text = element_text(size=12), axis.text.x = element_text(angle = 45, hjust = 1))
                
oecd.scatter
```

How we made graph 1:
We first specified that we wanted a new table called graph 1, we selected dataframe from the OECD and then used the pipe command to filter the data frame to only maintain the variables: Country, Year, Cost & LE. then we used again the pipe command to specify via filter that we only wanted the data from the year 2014. 

How we made the scatterplot:
We named the scatterplot oecd.scatter and used the command ggplot too make the graph, we specified that we wanted dataset graph 1 and we imported the axis with on the x-asis the Costs and on the Y-axis the variable LE. With theme we specified the shape to be round by shape = 16. We specified via stat_smooth that we wanted the line to be purple.

## Efficiency scores

The last assignment is to compute simple effiency scores. We will use the year 2014 for all countries (i.e. the dataframe "graph1").
Make a new dateframe called "eff" (using graph1). Please use the "pipe" to execute to following steps:
Select all the variables except the NA's (hint: use na.omit) and create a new column called "Ratio" with mutate. Ratio is defined as "Cost/LE" (cost per output).
Create a new column called "Score" with mutate. Score is the ratio per country divided by the lowest ratio in the dataframe.
To complete this dataframe order the dataset using "arrange" based on score (i.e. lowest Score first).


Finally make a bar chart using ggplot. Countries should be on the x-axis and "Score" on the y-axis. 



You can earn extra points if you are able to order the graph based on scores (lowest first, highest last), if you are able to make the x-axis readable, make the bar of the Netherlands red and the other bars blue and/or change the name of the y-axis to "Efficiency scores according to [your name]" 


The figure should like this:

![My Figure](eff.png)

```r
eff <- graph1 %>%
    filter(!is.na(LE)) %>%
    mutate(Ratio = Cost / LE) %>%
    mutate(Score = min(Ratio) / Ratio) %>%
    arrange(Score)
eff
```

```r
eff.bar <- ggplot(eff, aes(x = reorder(Country,Score), y = Score, fill= Country)) +
        geom_bar(stat = "identity") +
        theme(text = element_text(size=12), axis.text.x = element_text(angle = 45, hjust = 1))+
        scale_fill_manual(values = c(rep("blue",21),rep("red",1),rep("blue",12))) +
        guides(fill=FALSE) +
        xlab("Country") +
        ylab("Efficiency scores according to Kerkers & Korthout")

eff.bar
```

We made the new dataframe eff by altering dataframe called graph1 using pipe commands. Important to notice is that we created 2 new columns by the mutate command and later used the arrange function to order them from low to high scores. We defined the x-axis and y-axix and named them. With scale_fill_manual we colored the bar of the Netherlands red, by altering the value, the sum of these values is equal to the number of countries (34).
